```python
import numpy as np
from array import array

import plotly.express as px
```

```python
lambda_rate = np.random.uniform(0, 2000)
S = 10000 # Количество событий
lambda_rate
```

```python
def PoissonProcess(S, L):
    t = array('d')
    inter_times = array('d')
    sum = 0
    for i in range(S):
        inter = np.random.exponential(scale = 1/(lambda_rate), size=None)
        sum += inter
        t.append(sum)
        inter_times.append(inter)
    return t, inter_times
```

```python
event_times, inter_times = PoissonProcess(S, lambda_rate)
```

```python
fig = px.scatter(x=[1]*len(event_times), y=event_times, title=f'Poisson Process n S={S}, λ= 1/{lambda_rate}')
fig.update_traces(marker=dict(color='blue', size=5))  # уменьшение размера маркеров
fig.update_layout(width=300, height=700)  # уменьшение размера фигуры
fig.update_layout(xaxis_title='', yaxis_title='Time', xaxis=dict(showticklabels=False), yaxis=dict(showgrid=False, showticklabels=True))
fig.show()
```

```python
fig = px.histogram(np.array(event_times), nbins=1000, histnorm='density', title=f'Density Distribution of event Times  S={S}, λ= 1/{lambda_rate}')
fig.update_layout(xaxis_title='Event time', yaxis_title='Density', xaxis=dict(range=[0, max(event_times)]))
fig.show()
```

```python
fig = px.histogram(np.array(inter_times), nbins=1000, histnorm='density', title=f'Density Distribution of Inter Times S={S}, λ= 1/{lambda_rate}')
fig.update_layout(xaxis_title='Inter time', yaxis_title='Density', xaxis=dict(range=[0, max(inter_times)]))
fig.show()
```
